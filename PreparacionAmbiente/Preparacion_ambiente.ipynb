```python
%python
dbutils.widgets.removeAll()     
```

```python
%sql
CREATE WIDGET TEXT storage_account default="adlssmartdata1921"
```

```python
storage_account = dbutils.widgets.get("storage_account")
```

```python
%sql
CREATE EXTERNAL LOCATION IF NOT EXISTS `exlt-metastore`
URL 'abfss://metastore@${storage_account}.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL `credential`)
```

```python
%sql
CREATE EXTERNAL LOCATION IF NOT EXISTS `exlt-raw`
URL 'abfss://raw@${storage_account}.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL `credential`)
COMMENT 'External location raw';
```

```python
%sql
CREATE EXTERNAL LOCATION IF NOT EXISTS `extl-bronze`
URL 'abfss://bronze@${storage_account}.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL `credential`)
COMMENT 'External location bronze';
```

```python
%sql
CREATE EXTERNAL LOCATION IF NOT EXISTS `extl-silver`
URL 'abfss://silver@${storage_account}.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL `credential`)
COMMENT 'External location silver';
```

```python
%sql
CREATE EXTERNAL LOCATION IF NOT EXISTS `extl-golden`
URL 'abfss://golden@${storage_account}.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL `credential`)
COMMENT 'External location golden';
```

```python
%sql
DROP CATALOG IF EXISTS catalog_project CASCADE;
```

```python
%sql
CREATE CATALOG IF NOT EXISTS catalog_project
MANAGED LOCATION 'abfss://metastore@${storage_account}.dfs.core.windows.net/'
COMMENT 'Catalogo arquitectura madallion;'
```

```python
%sql
DROP SCHEMA IF EXISTS catalog_project.raw;
DROP SCHEMA IF EXISTS catalog_project.bronze;
DROP SCHEMA IF EXISTS catalog_project.silver;
DROP SCHEMA IF EXISTS catalog_project.golden;
```

```python
%python
dbutils.fs.rm(f"abfss://bronze@{storage_account}.dfs.core.windows.net/",True)
dbutils.fs.rm(f"abfss://silver@{storage_account}.dfs.core.windows.net/",True)
dbutils.fs.rm(f"abfss://golden@{storage_account}.dfs.core.windows.net/",True)
```

```python
%sql
CREATE SCHEMA IF NOT EXISTS catalog_project.raw;
CREATE SCHEMA IF NOT EXISTS catalog_project.bronze;
CREATE SCHEMA IF NOT EXISTS catalog_project.silver;
CREATE SCHEMA IF NOT EXISTS catalog_project.golden;
```

**Estructura tablas bronze**

```python
%sql
CREATE TABLE IF NOT EXISTS catalog_project.bronze.fileppe (
idTransaction integer,
cardNumber string,
account string,
dateppe date,
amount double,
ingestion_date timestamp
)
USING DELTA
LOCATION "abfss://bronze@${storage_account}.dfs.core.windows.net/fileppe"
```

```python

%sql
CREATE TABLE IF NOT EXISTS catalog_project.bronze.product (
idProducto integer,
descripcion string,
ingestion_date timestamp
)
USING DELTA
LOCATION "abfss://bronze@${storage_account}.dfs.core.windows.net/product"
```

```python
%sql
CREATE TABLE IF NOT EXISTS catalog_project.bronze.client (
idProducto integer,
idCliente integer,
Nombre string,
excenta BOOLEAN,
ingestion_date timestamp
) 
USING DELTA
LOCATION "abfss://bronze@${storage_account}.dfs.core.windows.net/client"
```

```python
%sql
CREATE TABLE IF NOT EXISTS catalog_project.bronze.accounts (
idProducto integer,
idCliente integer,
idNoCta string,
ingestion_date timestamp
)
USING DELTA
LOCATION "abfss://bronze@${storage_account}.dfs.core.windows.net/accounts"
```

**Estructura tablas silver**

```python
%sql
CREATE TABLE IF NOT EXISTS catalog_project.silver.totalbyproduct(
idproduct integer,
excenta BOOLEAN,
total DOUBLE
)
USING DELTA
LOCATION "abfss://silver@${storage_account}.dfs.core.windows.net/totalbyproduct"
```

```python
%sql
CREATE TABLE IF NOT EXISTS catalog_project.silver.cuentasbyclient(
idproduct integer,
descripcion string, 
nombre string,
totalcuentas integer
)
USING DELTA
LOCATION "abfss://silver@${storage_account}.dfs.core.windows.net/cuentasbyclient"
```

**Estructura tablas golden**

```python
%sql
CREATE TABLE IF NOT EXISTS catalog_project.golden.accountsbyclient(
idproduct integer,
descripcion string, 
nombre string,
totalcuentas integer
)
USING DELTA
LOCATION "abfss://golden@${storage_account}.dfs.core.windows.net/accountsbyclient"
```

```python
%sql
CREATE TABLE IF NOT EXISTS catalog_project.golden.totalbyproduct(
idproduct integer,
excenta BOOLEAN,
total DOUBLE
)
USING DELTA
LOCATION "abfss://golden@${storage_account}.dfs.core.windows.net/totalbyproduct"
```
