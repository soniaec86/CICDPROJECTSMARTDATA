```python

dbutils.widgets.removeAll()
```

```python

from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
dbutils.widgets.text("catalog", "catalog_project")
dbutils.widgets.text("schemaorigen", "silver")
dbutils.widgets.text("schemafinal", "golden")
```

```python

catalogo = dbutils.widgets.get("catalog")
esquema_source = dbutils.widgets.get("schemaorigen")
esquema_sink = dbutils.widgets.get("schemafinal")
```

```python
dftotalaccounts = spark.table(f"{catalogo}.{esquema_source}.cuentasbyclient")
```

```python
dftotalaccounts.write.mode("overwrite").insertInto(f"{catalogo}.{esquema_sink}.accountsbyclient")
```

```python
dftotalproduct = spark.table(f"{catalogo}.{esquema_source}.totalbyproduct")
```

```python
dftotalproduct.write.mode("overwrite").insertInto(f"{catalogo}.{esquema_sink}.totalbyproduct")
```
