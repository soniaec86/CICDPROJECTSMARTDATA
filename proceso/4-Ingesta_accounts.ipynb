```python
dbutils.widgets.removeAll()
```

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
dbutils.widgets.text("container", "raw")
dbutils.widgets.text("catalog", "catalog_project")
dbutils.widgets.text("schema", "bronze")
dbutils.widgets.text("storage-account", "adlssmartdata1921")
```

```python
container = dbutils.widgets.get("container")
catalog = dbutils.widgets.get("catalog")
schema = dbutils.widgets.get("schema")
storage_account = dbutils.widgets.get("storage-account")

file_path = f"abfss://raw@{storage_account}.dfs.core.windows.net/cuentas.csv"
```

```python
accounts = StructType(fields=[StructField("idProducto", IntegerType(), False),
                                  StructField("idCliente", IntegerType(), False),
                                  StructField("idNoCta", StringType(), False)
                                  ])
```

```python

accountsread = spark.read.option('header', True)\
                           .schema(accounts)\
                           .csv(file_path)
```

```python
accountsreaddt = accountsread.withColumn("ingestion_date", current_timestamp())

```

```python
df_accounts = accountsreaddt.select("idProducto",
                                                 "idCliente",
                                                 "idNoCta",
                                                 "ingestion_date"
                                                 )
```

```python

df_accounts.write.mode("overwrite").saveAsTable(f"{catalog}.{schema}.accounts")
```
